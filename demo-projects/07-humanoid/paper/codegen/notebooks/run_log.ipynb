# Run Log: 168-hour 4-site 32-iteration Camarade Swarm Run (v0.4.0)

Documents one full run of the PAT-NET-001 3-robot Unitree H2 EDU camarade swarm 24/7 AE response simulation. Read post-hoc for analysis.

```python
import json, pathlib
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

ROOT = pathlib.Path.cwd().parent
index_path = ROOT / 'data' / 'iterations' / 'index.jsonl'
iterations = [json.loads(line) for line in index_path.read_text().splitlines() if line.strip()]
print(f'Iterations loaded: {len(iterations)}')
```

## Iteration parameter table

```python
for it in iterations[:5]:
    print({k: it[k] for k in ['iteration_id', 'seed', 'llm_temperature', 'ae_arrival_jitter_seconds', 'response_time_variance', 'camarade_aggressiveness']})
```

## Per-site response time histogram

```python
p50s = [it['median_response_time_seconds'] for it in iterations]
p95s = [it['p95_response_time_seconds'] for it in iterations]
fig, ax = plt.subplots(figsize=(11, 8.5))
ax.hist(p50s, bins=12, alpha=0.5, label='p50', color='#1F3A68')
ax.hist(p95s, bins=12, alpha=0.5, label='p95', color='#8B2E3F')
ax.axvline(90, color='#2F6B3E', linestyle='--', label='90 s SLA')
ax.set_xlabel('Response time (seconds)')
ax.set_ylabel('Iteration count')
ax.set_title('Camarade swarm response time distribution across 32 iterations')
ax.legend()
fig.tight_layout()
fig.savefig(ROOT / 'reports' / 'figures' / 'run_log_response_time.png', dpi=300, facecolor='white')
plt.close(fig)
print('saved response_time figure')
```

## Camaraderie invariants pass rate

```python
rates = [it['camaraderie_invariants_pass_rate'] for it in iterations]
fig, ax = plt.subplots(figsize=(11, 8.5))
ax.bar(range(len(rates)), rates, color='#2E8B8B')
ax.set_xlabel('Iteration')
ax.set_ylabel('Pass rate')
ax.set_title('7 camaraderie invariants pass rate per iteration')
ax.set_ylim(0.95, 1.0)
fig.tight_layout()
fig.savefig(ROOT / 'reports' / 'figures' / 'run_log_camaraderie.png', dpi=300, facecolor='white')
plt.close(fig)
print('saved camaraderie figure')
```
